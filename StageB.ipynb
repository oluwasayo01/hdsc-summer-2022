```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.feature_selection import RFECV
```

```python
plt.figure(figsize=(20, 10))
```

```python
file = 'energydata_complete.csv'
```

```python
energy_data = pd.read_csv(file)
```

```python
energy_data
```

```python
energy_data.describe()
```

```python
energy_data.drop(['date', 'lights'], inplace=True, axis=1)
```

```python
energy_data
```

```python
energy_data.info()
```

```python
features = energy_data.iloc[:, 1:]
target = energy_data.iloc[:, 0]
```

```python
cols = features.columns
cols
```

```python
target
```

```python
data = features.values
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
X_test
```

```python
T2_index = 2
T6_index = 10
T2_train = X_train[T2_index]
T2_test = X_test[T2_index]
T2_train = X_train[T2_index]
T2_train = X_train[T2_index]
```

```python
transformed_df = pd.DataFrame(X_train, columns=cols)
transformed_df
```

```python
test_df = pd.DataFrame(X_test, columns=cols)
test_df
```

```python
x = transformed_df['T2'].values.reshape(-1, 1)
y = transformed_df['T6'].values
x_pred = test_df['T2'].values.reshape(-1, 1)
y_true = test_df['T6'].values
y_true
```

```python
reg = LinearRegression(n_jobs=-1)
reg.fit(x, y)
```

```python
y_pred = reg.predict(x_pred)
y_pred
```

```python
score = r2_score(y_true, y_pred)
score
```

```python
mae = mean_absolute_error(y_true, y_pred)
mae
```

```python
rss = np.sum(np.square(y_true - y_pred))
rss
```

```python
rmse = mean_squared_error(y_true, y_pred)
rmse
```

```python
x
```

```python
ridge = Ridge(alpha=0.4)
ridge.fit(x, y)
y_pred_ridge = ridge.predict(x_pred)
rmse_ridge = mean_squared_error(y_true, y_pred_ridge)
rmse_ridge
```

```python
rmse - rmse_ridge
```
